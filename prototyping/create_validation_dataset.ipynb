```python
import os
import random
import shutil
```

```python
original_path = "./dataset/DFWild-Cup"
```

```python
original_real_images = os.listdir(os.path.join(original_path, "train/real"))
original_fake_images = os.listdir(os.path.join(original_path, "train/fake"))
```

```python
valid_percent = 0.05
valid_size = int((len(original_fake_images) + len(original_real_images)) * valid_percent)
valid_size
```

```python
total_data = original_fake_images + original_real_images

for i in range(3):
    random.shuffle(total_data)

total_data[:10]
```

```python
valid_set_chosen = random.sample(total_data, valid_size)
valid_set_chosen[:10]
```

```python
reals = 0
fakes = 0

for path in valid_set_chosen:
    if "real" in path:
        reals += 1
    else:
        fakes += 1

print(f"Reals: {reals} || Fakes: {fakes}")
```

```python
for path in valid_set_chosen:
    if "real" in path:
        src_path = os.path.join(original_path, "train", "real", path)
        dest_path = os.path.join(original_path, "valid", "real", path)
    else:
        src_path = os.path.join(original_path, "train", "fake", path)
        dest_path = os.path.join(original_path, "valid", "fake", path)
    
    shutil.move(src_path, dest_path)
    print(f"Moved from {src_path} to {dest_path}")
```
